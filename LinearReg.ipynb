```python
import  pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#reading the data
df = pd.read_fwf('brain_body.txt')
```

```python
x_val = df[['Brain']]
y_val = df[['Body']]
```

```python
#Model training by using the linear_model object of sklearn
body_reg = linear_model.LinearRegression()
```

```python
#Fit the model on our x and y values from the dataframe
body_reg.fit(x_val, y_val)
```

<b>We have now obtained the line of best fit</b>

```python
#Plotting x, y value pair on a scatter plot
fig = plt.figure(figsize=(24,15))
plt.scatter(x_val, y_val)

#For every x value predict the y value
plt.plot(x_val, body_reg.predict(x_val))
```

# The entire plot

```python
#X-axis: Brain Weights
#Y-axis: Body Weights
plt.show()
```

# Partial zoom-in of the plot

```python
fig = plt.figure(figsize=(12,12))
#Plotting x, y value pair on a scatter plot
plt.scatter(x_val, y_val)

#For every x value predict the y value
plt.plot(x_val, body_reg.predict(x_val))

plt.xlim(-20.0, 500.0,)
plt.ylim(-20.0, 600.0)
plt.show()
```

# Calculating Mean Absolute Error 

```python
from sklearn.metrics import mean_absolute_error
print('Mean absolute error of this model is: ' , mean_absolute_error(y_val, body_reg.predict(x_val)))
```
