```python
import numpy as np
from scipy.optimize import minimize
from scipy.io import loadmat
from math import sqrt
import matplotlib.pyplot as plt
import random
import pickle
%matplotlib inline
```

```python
mat = loadmat('G:/Machine Learning/Assignment 1/face_all.mat')
```

```python
labels = mat['Labels'][0]
r = np.random.permutation(len(labels))
```

```python
fig = plt.figure(figsize=(24,12))
for i in range(20):
    plt.subplot(2,10,i+1)
    row = mat['Features'][r[i]]
    plt.imshow(np.reshape(row,((54,44))),cmap='gray')
    l = labels[r[i]]
    if l == 0:
        plt.title('No Glasses')
    else:
        plt.title('With Glasses')
    plt.axis('off')
```

```python
class0inds = np.where(labels == 0)[0]
class1inds = np.where(labels == 1)[0]
```

```python
r0 = np.random.permutation(len(class0inds))
r1 = np.random.permutation(len(class1inds))
```
