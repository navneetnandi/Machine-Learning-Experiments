```python
import numpy as np
from scipy.optimize import minimize
from scipy.io import loadmat
from math import sqrt
import matplotlib.pyplot as plt
import random
import pickle
%matplotlib inline
```

<b> Loading the mnist dataset </b>

```python
mat = loadmat('G:/Machine Learning/Assignment 1/mnist_all.mat')
```

```python
# For example train9 is all training images corresponding to the digit 9
train9 = mat.get('train9')
print(train9.shape)
```

```python
train8 = mat.get('train8')
print(train8.shape)
```

```python
#plot 100 random images from the digit '9' data set
s = random.sample(range(train8.shape[0]),100)
fig = plt.figure(figsize=(10,10))
for i in range(100):
    plt.subplot(10,10,i+1)
    row = train8[s[i],:]
    # note that each row is a flattened image
    # we first reshape it to a 28x28 matrix
    plt.imshow(np.reshape(row,((28,28))))
    plt.axis('off')
```

```python
fig = plt.figure(figsize=(12,12))
for i in range(10):
    trainx = mat.get('train'+str(i))
    # note the use of function random.sample to extract 10 rows randomly
    s = random.sample(range(trainx.shape[0]),10)
    for j in range(10):
        plt.subplot(10,10,10*i+j+1)
        row = trainx[s[i],:]
        # note that each row is a flattened image
        # we first reshape it to a 28x28 matrix
        plt.imshow(np.reshape(row,((28,28))))
        plt.axis('off')
```

<b>To check the total number of training examples you need to iterate over all trainx matrices</b>

```python
trainsize = 0
for i in range(10):
    m = mat.get('train'+str(i))
    trainsize = trainsize + m.shape[0] # m.shape will return the dimensions, ie. row X columns of the array m. So m.shape[0] gives us the total number of rows in m
print(trainsize)
```
